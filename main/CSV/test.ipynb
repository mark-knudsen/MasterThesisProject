```python
import math

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

n = len(x)

mean_x = sum(x) / n
mean_y = sum(y) / n

num = 0
denom_x = 0
denom_y = 0

for i in range(n):
    dx = x[i] - mean_x
    dy = y[i] - mean_y
    
    num += dx * dy
    denom_x += dx * dx
    denom_y += dy * dy

val = denom_x * denom_y

corr = num / math.sqrt(denom_x * denom_y)

print(corr)
print(val)
```
