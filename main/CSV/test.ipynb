```python
import pandas as pd
import numpy as np


df = pd.DataFrame(columns=["name", "age", "isCool", "score"], data=[["dan", 100, True, 20.2]])
```

```python
df
```

```python
new_data = pd.DataFrame([
    ["bob", 30, True, 85],
    ["lucas", 27, False, 60]
], columns=["name", "age", "isCool", "score"])

# Concatenating
df = pd.concat([df, new_data], ignore_index=True)
df
```

```python
df.to_csv("test.csv")
```
